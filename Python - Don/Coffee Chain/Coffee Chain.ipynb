```python
import pandas as pd
import numpy as np

df = pd.read_excel(r'C:\Users\Don.Rombaoa\Downloads\Python - Don\Coffee Chain\Coffee Chain.xlsx', sheet_name='Sales')
```

```python
# EXTRACT TIMESERIES (pd.to_datetime method)

# COPY Date & CONVERT TO DateTime
df['Date2'] = df['Date']
df['Date2'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
df['Date3'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d %H:%M:%S')

#Extract YEAR
df['Year'] = df['Date2'].dt.year


#EXTRACT MONTH
df['Month'] = df['Date2'].dt.month.astype(str).str.zfill(2)



#Extract Day of Month
df['Day'] = df['Date2'].dt.day.astype(str).str.zfill(2)



#Extract Day of Week (0 Monday to 6 Sunday)
df['Weekday'] = df['Date2'].dt.dayofweek



#Extract Day of Week (Monday to Sunday)
df['DayName'] = df['Date2'].dt.day_name()



#Extract WEEK
df['Week'] = df['Date2'].dt.weekofyear

df['Week2'] = df['Date2'].dt.isocalendar().week



#EXTRACT HOUR
df['Hour'] = df['Date2'].dt.hour



#EXTRACT MINUTES
df['Minutes'] = df['Date2'].dt.minute



#EXTRACT SECONDS
df['Seconds'] = df['Date2'].dt.second

#############################
#ALTERNATIVE FORMAT - SLASH - MINS & HOURS  format = '%d/%m/%Y %H:%M:%S')

#df['DayofWeek2'] = df['Date_Alt'].dt.day_name()
#df


#df['Flag_Weekend'] = df[Day]

#df['date/time'] = pd.to_datetime(df['date/time'], format = '%d/%m/%Y %H:%M:%S')
```

```python
# Extract - Time

# Extract Year
df['Year'] = pd.DatetimeIndex(df['Date']).year

#Extract Month
df['Month'] = pd.DatetimeIndex(df['Date']).month

#Extract Day (note for this case we added leading zeros)
df['DayofMonth'] = pd.DatetimeIndex(df['Date']).day.astype(str).str.zfill(2)
df


#Extract Day of Week (0 Monday - 6 Sunday)
df['DayofWeek1'] = pd.DatetimeIndex(df['Date']).dayofweek.astype(str).str.zfill(2)
```

```python
df = df.drop(['year'], axis=1)
df
```

```python
# CONCATENATE - INCLUDES LEADING ZEROS FOR MONTHS
df['Year_Month'] = df['Year'].map(str) + '-' + df['Month'].astype(str).str.zfill(2)
df
```

```python
# DROP COLUMNS

#Drop Multi-Columns
df.drop(['year','month'], axis=1)
```

```python
# VISUALISE KPI ACROSS YEARS-MONTHS TO CHECK FOR TRENDS AND DATA GAPS

df[['Year_Month','Sales']].groupby(by = ['Year_Month']).sum('Sales').plot(figsize=(10,6), fontsize=12, title='Monthly Sales of Products')
```
